# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Dataset

```python
data = pd.read_csv('../data/raw/kiva_loans.csv')
```

```python
data.head(10)  # Display the first 10 rows of the dataset
```

## Sector Analysis

### Most Popular

```python
data["sector"].value_counts().head(10).plot(kind='barh', figsize=(10, 6))
plt.title('Most Popular Sectors in Kiva Loans')
plt.xlabel('Number of Loans')
plt.ylabel('Sector')
plt.show()
```

The most popular secion is Agriculture..

### Most Funded Sector 

```python
sns.barplot(x='sector', y='funded_amount', data=data, palette='Set2')
plt.title('Distribution of Funded Amount by Sector')
plt.xlabel('Sector')
plt.ylabel('Funded Amount')
plt.xticks(rotation=45)
plt.show()
```

The most well-funded sector is Manufacturing. 

## Distribution of Funded and Loan Amount

```python
data["funded_amount"].plot(kind='hist', bins=100, figsize=(10, 6), alpha=0.7)
plt.title('Distribution of Funded Amount')
plt.xlabel('Funded Amount')
plt.ylabel('Frequency')
plt.xlim(0, 20000)  # Adjust x-axis limit for better visibility
plt.show()

data["loan_amount"].plot(kind='hist', bins=100, figsize=(10, 6), alpha=0.7)
plt.title('Distribution of Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.xlim(0, 20000)  # Adjust x-axis limit for better visibility
plt.show()
```

The distribution between loan amount and fund amount are almost identical. 

## Country

### Most Popular

```python
data['country'].value_counts().head(10).plot(kind='barh', figsize=(10, 6))
plt.title('Most Popular Countries for Kiva Loans')
plt.xlabel('Number of Loans')
plt.ylabel('Country')
plt.show()
```

The country with the most Kiva loans is Philippines. 

### Most Funded

```python
plt.figure(figsize=(20, 6))
sns.barplot(x='country', y='funded_amount', data=data, palette='Set2')
plt.title('Distribution of Funded Amount by Country')
plt.xlabel('Country')
plt.ylabel('Funded Amount')
plt.xticks(rotation=45)
plt.show()
```

The most well-funded country is Cte d'Ivoire. 

## Common Intervals

```python
data['repayment_interval'].value_counts().plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Repayment Intervals')
plt.xlabel('Repayment Interval')
plt.ylabel('Number of Loans')
plt.xticks(rotation=45)
plt.show()
```

Monthly and irregular are the most common repayment intervals. 

## Terms in Months Distribution

```python
data['term_in_months'].plot(kind='hist', bins=30, figsize=(10, 6), alpha=0.7)
plt.title('Distribution of Loan Terms in Months')
plt.xlabel('Term in Months')
plt.ylabel('Frequency')
plt.xlim(0, 60)  # Adjust x-axis limit for better visibility
plt.show()
```

Most loan terms are under one year with the average around 10-12.
